```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading data
data = pd.read_csv('Dataset/HeartDisease.csv')
data.head()
```

```python
sns.heatmap(data.drop(columns='Heart_Disease').corr())
```

```python
labels = data.columns.tolist()
def func(text):
  text = text.replace('_', " ")
  return text

for label in labels:
  sns.countplot(data, x = label)
  plt.xlabel(func(label))
  plt.ylabel('Frequency')
  plt.title('Frequency of ' + func(label))
  plt.show()
```
